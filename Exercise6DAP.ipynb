there seems to be some empty records, let's check them and drop remove them using dropna()  
we will use ...dropna(how='all') to get rid of all empty records

```python
df2 = df
df2.dropna(how='all', inplace=True)
#df2
```

let's remove duplicate entries 'full_name'

```python
#df2.duplicated(['full_name'])
df2 = df
df2.drop_duplicates(['full_name'], inplace=True)
```

Now we remove all records, where at least one field has no value

Removing the whole record, if only the id is missing? That would not be wise.  
Note: in this demo dataframe we have only one entry with missing emailaddress and one with missing gender  
We could replace missing gender by cecking field first_name, but we don't have a name database to look for name<->gender

```python
df2['full_name'].dropna(inplace=True)
df2['first_name'].dropna(inplace=True)
df2['last_name'].dropna(inplace=True)
df2['email'].dropna(inplace=True)
df2['gender'].dropna(inplace=True)
df2['age'].dropna(inplace=True)
```
